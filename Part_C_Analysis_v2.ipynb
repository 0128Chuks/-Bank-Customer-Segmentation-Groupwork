# Statistics for 5 locations

```python
# import stats libraries
from scipy import stats
```

```python
# Transform the data to induce a more symmetrical form
customers_fix_1 = pd.DataFrame()
customers_fix_1["Recency"] = stats.boxcox(customers_1['Recency'])[0]
customers_fix_1["Frequency"] = stats.boxcox(customers_1['Frequency'])[0]
customers_fix_1["MonetaryValue"] = stats.boxcox(customers_1['Monetary'])[0]

customers_fix_2 = pd.DataFrame()
customers_fix_2["Recency"] = stats.boxcox(customers_2['Recency'])[0]
customers_fix_2["Frequency"] = stats.boxcox(customers_2['Frequency'])[0]
customers_fix_2["MonetaryValue"] = stats.boxcox(customers_2['Monetary'])[0]

customers_fix_3 = pd.DataFrame()
customers_fix_3["Recency"] = stats.boxcox(customers_3['Recency'])[0]
customers_fix_3["Frequency"] = stats.boxcox(customers_3['Frequency'])[0]
customers_fix_3["MonetaryValue"] = stats.boxcox(customers_3['Monetary'])[0]

customers_fix_4 = pd.DataFrame()
customers_fix_4["Recency"] = stats.boxcox(customers_4['Recency'])[0]
customers_fix_4["Frequency"] = stats.boxcox(customers_4['Frequency'])[0]
customers_fix_4["MonetaryValue"] = stats.boxcox(customers_4['Monetary'])[0]

customers_fix_5 = pd.DataFrame()
customers_fix_5["Recency"] = stats.boxcox(customers_5['Recency'])[0]
customers_fix_5["Frequency"] = stats.boxcox(customers_5['Frequency'])[0]
customers_fix_5["MonetaryValue"] = stats.boxcox(customers_5['Monetary'])[0]
```

```python
# Plot Recency graphs to show that box-cox transformation worked
plt.figure(figsize=(17,14))

plt.subplot(5, 1, 1); sns.distplot(customers_fix_1['Recency'])

plt.subplot(5, 1, 2); sns.distplot(customers_fix_2['Recency'])

plt.subplot(5, 1, 3); sns.distplot(customers_fix_3['Recency'])

plt.subplot(5, 1, 4); sns.distplot(customers_fix_4['Recency'])

plt.subplot(5, 1, 5); sns.distplot(customers_fix_5['Recency'])

plt.show()
```

```python
# Plot Frequency graphs to show that box-cox transformation worked
plt.figure(figsize=(17,14))

plt.subplot(5, 1, 1); sns.distplot(customers_fix_1['Frequency'])

plt.subplot(5, 1, 2); sns.distplot(customers_fix_2['Frequency'])

plt.subplot(5, 1, 3); sns.distplot(customers_fix_3['Frequency'])

plt.subplot(5, 1, 4); sns.distplot(customers_fix_4['Frequency'])

plt.subplot(5, 1, 5); sns.distplot(customers_fix_5['Frequency'])

plt.show()
```

# Apply K Means methodology

```python
# import k means
from sklearn.cluster import KMeans
```

```python
# import libraries for more in depth analysis
from yellowbrick.cluster import KElbowVisualizer
from sklearn.cluster import KMeans
```

# Analysis of Location 1

```python
# apply elbow method to determine number of categories - location 1

sse = {}
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=32)
    kmeans.fit(customers_1_normalized)
    sse[k] = kmeans.inertia_ # SSE to closest cluster centroid
plt.title('The Elbow Method - Delhi')
plt.xlabel('k')
plt.ylabel('SSE')
sns.pointplot(x=list(sse.keys()), y=list(sse.values()))
plt.show()
```

```python
model_1 = KMeans()
visualizer = KElbowVisualizer(model_1, k=(1,12))
visualizer.fit(customers_1_normalized)
visualizer.show()
```

```python
model_1 = KMeans(n_clusters=4, random_state=32) #utilizing K from above
model_1.fit(customers_1_normalized)
model_1.labels_.shape
customers_1["Cluster"] = model_1.labels_
customers_1.groupby('Cluster').agg({
    'Recency':'mean',
    'Frequency':'mean',
    'Monetary':['mean', 'count']}).round(2)
f, ax = plt.subplots(figsize=(25, 5))
ax = sns.countplot(x="Cluster", data=customers_1)
customers_1.groupby(['Cluster']).count()
```

```python
customers_1["Cluster"] = model_1.labels_
customers_1.groupby('Cluster').agg({
    'Recency':'mean',
    'Frequency':'mean',
    'Monetary':['mean', 'count']}).round(2)
```

```python
# Create the dataframe
df_1_normalized = pd.DataFrame(customers_1_normalized, columns=['Recency', 'Frequency', 'Monetary'])
df_1_normalized['ID'] = customers_1.index
df_1_normalized['Cluster'] = model_1.labels_

# Melt The Data
df_nor_melt_1 = pd.melt(df_1_normalized.reset_index(),
                      id_vars=['ID', 'Cluster'],
                      value_vars=['Recency','Frequency','Monetary'],
                      var_name='Attribute',
                      value_name='Value')
df_nor_melt_1.head()

# Visualize data
sns.lineplot(x = 'Attribute', y= 'Value', hue='Cluster', data=df_nor_melt_1).set_title("Location 1 - Delhi")
```
